```python
import pandas as pd
import numpy as np

df = pd.read_csv("../data/cleaned_telco_churn.csv")
```

encoding target variable

```python
df['Churn'] = df['Churn'].map({'Yes': 1, 'No': 0})
#1-> Churned, 0-> Not Churned
```

Encode Binary Categorical Features

```python
binary_cols = [
    'Partner', 'Dependents', 'PhoneService',
    'PaperlessBilling'
]

for col in binary_cols:
    df[col] = df[col].map({'Yes': 1, 'No': 0})
```

One-Hot Encode Multi-Category Features

```python
# One-Hot Encoding for categorical variables with more than two categories
df = pd.get_dummies(
    df,
    columns=[
        'gender', 'MultipleLines', 'InternetService',
        'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
        'TechSupport', 'StreamingTV', 'StreamingMovies',
        'Contract', 'PaymentMethod'
    ],
    drop_first=True
)
```

Feature Scaling (Prepare for ML)

Tree models donâ€™t need scaling,
but Logistic Regression does.

We will prepare for both.

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

num_cols = ['tenure', 'MonthlyCharges', 'TotalCharges']
df[num_cols] = scaler.fit_transform(df[num_cols])
import joblib

joblib.dump(scaler, "../models/scaler.pkl")
```

```python
#seperate features and target variable
X = df.drop('Churn', axis=1)
y = df['Churn']
```

```python
# Save the processed features and target variable/ featured engineered data
X.to_csv("../data/X_features.csv", index=False)
y.to_csv("../data/y_target.csv", index=False)
```
